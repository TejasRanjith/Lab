# 2. Mean, Median, Mode

```python
def mean(l):
    s=0
    for elem in l:
        s+=elem
    return s/len(l)

def sorter(l):
    for i in range(len(l)-1):
        for j in range(len(l)-1):
            if l[j] > l[j+1]:
                temp = l[j]
                l[j] = l[j+1]
                l[j+1] = temp
    return l

def median(l):
    l,n = sorter(l),len(l)-1
    if n%2 == 0:
        return l[n//2]
    else:
        return (l[n//2] + l[((n+2)//2)])/2

def mode(l):
    d,result=dict(),[]
    for elem in l:
        if elem in d:
            d[elem]+=1
        else:
            d[elem]=1
    for key in d:
        if d[key] == max(d.values()):
            if max(d.values()) != 1:
                result.append(key)
            else:
                result = [0]
    return result

def variance(l):
    xbar = mean(l)
    newl = [(x-xbar)**2 for x in l]
    return sum(newl)/(len(l)-1)

l = [int(x) for x in input("Enter the elements accordingly : ").split()]
print("Mean of the list: ",mean(l))
print("Median of the list: ",median(l))
print("Mode of the list: ",*mode(l))
print("Variance of the list: ",variance(l))
print("Standard Deviation of list: ",variance(l)**(1/2))
```

```python
import re

def validator(password):
    errors = []
    if len(password)<=8:
        errors.append("Does not contain 8 or more characters...")
    if not re.search("[A-Z]",password):
        errors.append("Does not contain uppercase letters...")
    if not re.search("[a-z]",password):
        errors.append("Does not contain lowercase letters...")
    if not re.search("[0-9]",password):
        errors.append("Does not contain digits...")
    if not re.search("[!@#$%^&*]",password):
        errors.append("Does not contain special characters...")
    return errors

errors = validator(input("Enter the password to be validated : "))
for error in errors:
    print(error)
if len(errors) == 0:
    print("Valid Password")
else:
    print("Invalid Password")
```

```python
def timeOfDay(n):
    if n in range(1,4):
        return "Late Night"
    elif n in range(4,7):
        return "Early Morning"
    elif n in range(7,12):
        return "Morning"
    elif n in range(12,15):
        return "After Noon"
    elif n in range(15,18):
        return "Evening"
    elif n in range(18,25) or n==0:
        return "Night"
```

```python
df['timeofday'] = df['hour'].apply(lambda x:timeOfDay(x))
```

```python
def monthToSeasons(n):
    if n in [9,10,11]:
        return "Spring"
    elif n in [12,1,2]:
        return "Summer"
    elif n in [3,4,5]:
        return "Autumn"
    elif n in [6,7,8]:
        return "Winter"
```

```python
df['seasons'] = df['month'].apply(lambda x:monthToSeasons(x))
```
